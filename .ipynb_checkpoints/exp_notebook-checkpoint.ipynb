```python
import os
import numpy as np
from glob import glob
from PIL import Image
from matplotlib import pyplot as plt
```

```python
files = glob("output_dir/*.png")
base_dir = "/home/saad/DataDrive1/Dataset/H36M/Cropped_Data/images/valid/"

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
bx = fig.add_subplot(122)
plt.ion()

for filepath in files:
    actual_filename = filepath.split("/")[1].split(".")[0]
    gt_filepath = os.path.join(base_dir, f"{actual_filename}.jpg")
    with Image.open(filepath) as pred, Image.open(gt_filepath) as gt:
        ax.imshow(gt)
        bx.imshow(pred)
        
        fig.canvas.draw()
        fig.canvas.flush_events()
        plt.pause(0.0001)
        ax.clear()
        bx.clear()

plt.show()
```
